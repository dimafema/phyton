```python
class Autor: # clase Autor
    def __init__(self,nombre,apellidos):
        self.Nombre = nombre  # Compuesta por dos atributos "Nombre, Apellidos"
        self.Apellidos = apellidos
    def MostrarAutor(self): # Creamos el método que muestra por pantalla los atributos "Nombre, Apellidos"
        print("Autor: ",self.Nombre," ",self.Apellidos)

class Libro: # clase Libro
    def __init__(self,titulo,isbn):
        self.Titulo = titulo # se componen de dos atributos "Titulo, ISNB"
        self.ISBN = isbn
    def AñadirAutor(self,autor): # método que añade autor
        self.Autor = autor # este otro atributo Autor, es de tipo Autor, osea coge los atributo de la clase autor
    def MostrarLibro(self): # método que muestra el libro
        print("------Libro------")
        print("Título: ",self.Titulo)
        print("ISBN: ",self.ISBN)
        self.Autor.MostrarAutor()# se puede mostrar el autor, pues podemos apuntar al metodo de la clase Autor, por que tenemos el atributo Autor tipo Autor
        print("----------------")
    def ObtenerTitulo(self): # método devuelve el título del libro
        return self.Titulo;

class Biblioteca:# clase Biblioteca
    def __init__(self): 
        self.ListaLibros = []# Con un atributo "listaLibros" que es de tipo lista
    def NumeroLibros(self): # método que devuelve el número de libro que tenemos en la lista
        return len(self.ListaLibros)# retornamos con el uso del la condición "len" para que cuente la listaLibros
    def AñadirLibro(self,libro):# método añade libro  a la lista
        self.ListaLibros = self.ListaLibros + [libro] # pasamos el libro por parámetro
    def MostrarBiblioteca(self):# método que muestra en pantalla la lista de libros
        print("###############################")
        for item in self.ListaLibros:
            item.MostrarLibro()
        print("###############################")
    def BorrarLibro(self,titulo):# Elimina un item de la lista de libro a partir del título de este
        encontrado = False # inicializamos una variable en False
        posicionaBorrar = -1 # inicializamos una variable en -1
        for item in self.ListarLibros: # mientras haya item de libro recorremos lista
            posicionaBorrar += 1 # sumamos 1 a la variable
            if item.ObtenerTitulo() == titulo: # si el valor del parametro es identico al valor del resultado que me da el método ObtenerTitulo
                encontrado = True # cambiamos el valor de la variable a True
            break # salimos de ciclo if
        if encontrado: # como hemos encontrado el libro (True)
            del self.ListarLibros[posicionaBorrar] # borramos por su indice
            print("¡Libro borrado correctamente!") # mostramos que hemos borrado el libro
        else:# sino 
            print("¡Libro no encontrado") # imprimimos que no encontramos el libro

def MostrarMenu():# función mostras menú
    print("""Menu
    1) Añadir libro a la biblioteca
    2) Mostrar biblioteca
    3) Borrar libro
    4) ¿Número de libros?
    5) Salir""")

def AñadirLibroABiblioteca(biblioteca):# función añadir libro
    titulo = input("Introduzca del título del libro") # damos una entrada título del libro
    isbn = input("Introduzca el isbn: ") # damos una entrada del código isbn
    autornombre = input("introduzca el nombre del autor: ")# damos una entrada nombre del autor
    autorapellidos = input("Introduzca los apellidos del autor: ")# damos una entrada de los apellidos del autor
    autor = Autor(autornombre,autorapellidos)# variable autor damos parámetros de las variables input y agregamos a la clase Autor(nombre, apellido)
    libro = Libro(titulo, isbn)# variable libro damos parámetros de las variables input y agremamos a la clase libro
    libro.AñadirAutor(autor)# añadimos autor a la clase libro
    biblioteca.AñadirLibro(libro)# añadimos a la clase biblioteca el libro
    return biblioteca# devuelve clase biblioteca
def MostrarBiblioteca(biblioteca):
    biblioteca.MostrarBiblioteca()# llamamos a la función mostrar lista de libro de la biblioteca de la clase biblioteca

def BorrarLibro(biblioteca):
    titulo = input("Introduzca el título del libro a borrar: ")# damos un input para introducir el título
    biblioteca.BorrarLibro(titulo)# llamamos al método borrar libro de la clase biblioteca

def NumeroLibros(biblioteca):
    print("El número de libros en la biblioteca es: ",biblioteca.NumeroLibros())# se muestra el número de libros que tiene la biblioteca

fin = False # iniciamos la variable fin en False
biblioteca = Biblioteca() # asignamos a la variable la clase biblioteca

while not fin:# mietra fin sea a True, muestra el menú
    MostrarMenu()# llamamos al método MostrarMenu
    opcion = int(input("Seleccione opción: "))# entre un número del menú
    if(opcion == 1):
        biblioteca = AñadirLibroABiblioteca(biblioteca)# llamamos al método añadir título a la biblioteca
    elif(opcion == 2):
        MostrarBiblioteca(biblioteca)# llamamos al método mostrar libro de la biblioteca
    elif(opcion == 3):
        BorrarLibro(biblioteca)# llamamos al método borrar libro
    elif(opcion == 4):
        NumeroLibros(biblioteca)# llamamos al método leer el número de libro de la biblioteca
    elif(opcion == 5):
        fin = True# conertimos la variable fin en True y salimos del ciclo 

print("¡Adios!")# imprimimos la despedida
```
